```python
import sqlite3
import pandas as pd
import os

DB_PATH = "/Users/vinhlam/PycharmProjects/smartphones-review-analyzer/data/processed/reviews.db"

conn = sqlite3.connect(DB_PATH)
cursor = conn.cursor()
print(f"Database created at: {DB_PATH}")
```

```python
BASE = "/Users/vinhlam/PycharmProjects/smartphones-review-analyzer/data/processed/"

df_main = pd.read_csv(BASE + "master_topics.csv", low_memory=False)
df_brand = pd.read_csv(BASE + "summary_brand.csv")
df_brand_topic = pd.read_csv(BASE + "summary_brand_topic.csv")
df_brand_price = pd.read_csv(BASE + "summary_brand_price.csv")
df_features = pd.read_csv(BASE + "summary_features.csv")

df_main.to_sql('reviews', conn, if_exists='replace', index=False)
df_brand.to_sql('brand_summary', conn, if_exists='replace', index=False)
df_brand_topic.to_sql('brand_topic_summary', conn, if_exists='replace', index=False)
df_brand_price.to_sql('brand_price_summary', conn, if_exists='replace', index=False)
df_features.to_sql('feature_summary', conn, if_exists='replace', index=False)

print("Tables created in database:")
tables = cursor.execute("SELECT name FROM sqlite_master WHERE type='table'").fetchall()
for t in tables:
    count = cursor.execute(f"SELECT COUNT(*) FROM {t[0]}").fetchone()[0]
    print(f"  {t[0]}: {count:,} rows")
```
